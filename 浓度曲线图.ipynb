```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('con_data.xlsx')
```

```python
data3=pd.read_csv('history.csv')
```

```python
data3
```

```python

ax = plt.axes()
ax.plot(data3['loss'],label='Loss',color='red') #实验数据
ax.plot(data3['val_loss'],label='Val_loss') #模拟数据
plt.xlabel('Epochs',fontproperties='Times New Roman',fontsize=18) #x轴名
plt.ylabel('Loss',fontproperties='Times New Roman',fontsize=18) #y轴名
ax.grid( linestyle=':',which='major',linewidth=1,alpha=0.5) #网格线
ax.tick_params(axis='both',which='both',direction='in') #刻度方向
#plt.title(f'Sensor {dic[i]}',fontproperties='Times New Roman',fontsize=20) #标题
plt.minorticks_on() #打开次要刻度
plt.xticks(fontproperties='Times New Roman',fontsize=16) #x轴刻度
plt.yticks(fontproperties='Times New Roman',fontsize=16)# y轴刻度
ax.legend(prop={'size':15,'family':'Times New Roman'},loc=1) #图例
plt.show()
#plt.savefig(f'{i}.png')
plt.clf()
```

```python
for i in ['202','106','103','101','104','204']: #需要画的传感器编号
    dic={'202':'Ⅻ','106':'Ⅵ','103':'Ⅴ','101':'Ⅰ','104':'Ⅹ','204':'Ⅳ'}
    plt.rcParams['figure.figsize'] = (6.0, 4.0)
    ax = plt.axes()
    ax.plot(data[i+'et'][0:400],data[i+'ec'][0:400],label='Exp',color='red') #实验数据
    ax.plot(data[i+'t'][0:2500],data[i+'c'][0:2500],label='CFD') #模拟数据
    plt.xlabel('Time(s)',fontproperties='Times New Roman',fontsize=18) #x轴名
    plt.ylabel('He concentration',fontproperties='Times New Roman',fontsize=18) #y轴名
    ax.grid( linestyle=':',which='major',linewidth=1,alpha=0.5) #网格线
    ax.tick_params(axis='both',which='both',direction='in') #刻度方向
    plt.title(f'Sensor {dic[i]}',fontproperties='Times New Roman',fontsize=20) #标题
    plt.minorticks_on() #打开次要刻度
    plt.xticks(fontproperties='Times New Roman',fontsize=16) #x轴刻度
    plt.yticks(fontproperties='Times New Roman',fontsize=16)# y轴刻度
    ax.legend(prop={'size':15,'family':'Times New Roman'},loc=2) #图例
    plt.show()
    #plt.savefig(f'{i}.png')
    plt.clf()
```
