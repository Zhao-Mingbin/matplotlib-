```python
#x
import matplotlib.pyplot as plt
import numpy as np
def  f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
a = np.arange(0,5,0.02)
plt.subplot(211)
plt.plot(a,f(a))
plt.subplot(212)
plt.plot(a,np.cos(2*np.pi*a),color='r', linestyle='--', marker='.') #plt.plot(a,np.cos(2*np.pi*a),'r--.') 
plt.savefig('1')
plt.show()

```

```python
# 
matplotlib.rcParams['font.family']='FangSong'
matplotlib.rcParams['axes.unicode_minus']=False
matplotlib.rcParams['font.size']=13
"""

explode
labels:
labeldistance:1.11.1
autopct
shadow
startangle0
pctdistance
"""
labels = ['','','','']
sizes = [14.55,30.45,45,10]
colors = ['red','yellow','lightskyblue','green']
explode = (0,0.1,0,0)
# 
plt.figure(figsize=(7,4),dpi=80)
# 
plt.pie(sizes,labels=labels,colors=colors,explode=explode,labeldistance = 1.1,pctdistance = 0.6,autopct='%.2f%%',shadow=False,startangle=90)
plt.axis('equal') #xy
plt.legend(loc='best')
plt.savefig('')
plt.show()
```

```python
np.random.seed(0)
mu,sigma = 100,20
data = np.random.normal(mu,sigma,size=100)
"""

data:
bins:10
normed:0normed=1
facecolor:
edgecolor:
alpha:
histtype::'bar', 'barstacked','step', 'stepfilled'
"""
plt.hist(data,bins=20,normed=0,histtype='barstacked',facecolor='g',edgecolor="black",alpha=0.6)
# 
plt.xlabel("",fontproperties='SimHei',fontsize=12)
# 
plt.ylabel("/",fontproperties='SimHei',fontsize=12)
# 
plt.title('/',fontproperties='SimHei')
plt.savefig('')
plt.show()
```

```python
matplotlib.rcParams['font.family']='SimHei'
matplotlib.rcParams['axes.unicode_minus']=False

label_list=['2015','2016','2017','2018'] #
num_list1=[20,30,15,35] # 1
num_list2=[15,30,40,20] #2
x = range(len(num_list1))
"""

x:
height:
width:0.8
label:legend
"""
plt.figure(figsize=(8,5),dpi=100)
rects1 = plt.bar(x,height=num_list1,width=0.4,alpha=0.7,color='red',label='')
rects2 = plt.bar([i+0.4 for i in x],height=num_list2,width=0.4,alpha=0.7,color='blue',label='')
plt.ylim(0,50)
plt.ylabel('')
"""
x


"""
plt.xticks([index+0.2 for index in x],label_list)
plt.xlabel('')
plt.title('')
#
for rect in rects1:
    height = rect.get_height()
    plt.text(rect.get_x()+rect.get_width()/2,height+1,str(height),ha='center',va='bottom')
for rect in rects2:
    height = rect.get_height()
    plt.text(rect.get_x()+rect.get_width()/2,height+1,str(height),ha='center',va='bottom')
plt.legend()
plt.savefig('')
plt.show()
```

```python
matplotlib.rcParams['font.family'] = 'SimHei'
matplotlib.rcParams['axes.unicode_minus'] = False

price = [39.5,44.2,36.7,38.9,33.33]
"""
barh
y
x
"""
plt.figure(figsize=(8,5),dpi=80)
_price = plt.barh(range(5),price,height=0.6,color='cyan',alpha=0.7)
plt.yticks(range(5),['','','','',''])
plt.xlim(30,47)
plt.xlabel('')
plt.title('')
plt.grid(alpha=0.2)
#
for x in _price:
    width = x.get_width()
    plt.text(width+1,x.get_y()+x.get_height()/2,str(width),ha='center',va='center')
plt.savefig('')
plt.show()
```
